## Spark Operations

### Identify Main Categories with Only One Entry

Steps:
1. Group by category and count the number of entries for each
2. Filter to get categories that have only one entry
3. Collect the results into a list
4. Print the list of categories with only one entry


```python
from pyspark.sql import functions as F

category_counts = df.groupBy('category').count()
single_entry_categories = category_counts.filter(F.col('count') == 1).select('category') 
single_entry_categories_list2 = [row.category for row in single_entry_categories.collect()]

print(single_entry_categories_list2)
```

#### How many values are in single_entry_categories_list

```python
len(single_entry_categories_list2)
```

#### Filter out the rows that have a main category in the single_entry_categories_list

```python
df = df.filter(~df['category'].isin(single_entry_categories_list2))

df.show()
```

### Categories with one entry
1. Group by main_category and count the number of entries for each
2. Filter to get main categories that have only one entry
3. Collect the results into a list

```python
from pyspark.sql import functions as F

category_counts = df.groupBy('main_category').count()
single_entry_categories = category_counts.filter(F.col('count') == 1).select('main_category')
single_entry_categories_list = [row.main_category for row in single_entry_categories.collect()]

print(single_entry_categories_list)
```

#### How many values are in single_entry_categories_list

```python
len(single_entry_categories_list)
```

#### Filter out the rows that have a main category in the single_entry_categories_list


```python
df = df.filter(~df['main_category'].isin(single_entry_categories_list))

df.show()
```

## All the rows removed are less than 1% of the dataset

```python
# Average number of backers for each main category in descending order
df.groupBy('main_category').agg(avg('backers')).orderBy('avg(backers)', ascending = False).show(15)
```

## Visualization

```python
df.groupBy('main_category').agg(avg('backers')).orderBy('avg(backers)', ascending = False).toPandas().plot(kind='bar',
x='main_category', y='avg(backers)', figsize=(10,5), colormap='coolwarm')
plt.title('Average Number of Backers for Each Main Category')
plt.ylabel('Average Number of Backers')
plt.xlabel('Main Category')
plt.show()
```

As we can see Games are the most attractive category for backers

```python
df.groupBy('main_category').agg(avg('usd_pledged_real')).orderBy('avg(usd_pledged_real)', ascending = False).show(15)
```

```python
category_names = df.select('main_category').distinct().toPandas()['main_category'].tolist()
df_grouped = df.groupBy('main_category').agg(avg('usd_pledged_real')).orderBy('avg(usd_pledged_real)', ascending = False).toPandas()
df_grouped['main_category'] = df_grouped['main_category']
df_grouped.plot.pie(y='avg(usd_pledged_real)', labels=df_grouped['main_category'], autopct='%1.1f%%', figsize=(10, 10), legend=False)
plt.title('Average Pledged Amount for Each Main Category')
plt.ylabel('')
plt.show()
```

As we can see from the query results, Design has the most pledged money being put in for funding


```python
df.groupBy('main_category').agg(avg('backers'), avg('usd_pledged_real')).orderBy('avg(backers)', ascending = False).show(15)
```

### Mixing both 

```python
df_grouped = df.groupBy('main_category').agg(avg('backers'), avg('usd_pledged_real')).orderBy('avg(backers)', ascending = False).toPandas()
plt.figure(figsize=(10, 5))
plt.scatter(df_grouped['main_category'], df_grouped['avg(backers)'], s=df_grouped['avg(usd_pledged_real)'], alpha=0.5, color='darkblue')
plt.title('Average Number of Backers and Pledged Amount for Each Main Category')
plt.xlabel('Main Category')
plt.ylabel('Average Number of Backers')
plt.xticks(rotation=45)
plt.show()
```
